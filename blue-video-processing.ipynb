```python
import cv2
import numpy as np
import math

filename = "test.avi"

video = cv2.VideoCapture(filename)
history = []

def write(image, txt, lineno=1):
    cv2.putText(image, txt, (10,340+40*lineno), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (209, 80, 0, 255), 3)

def get_dy(clean_image, drawable_image):
    # gaussian blur
    blurred = cv2.GaussianBlur(clean_image,(25,25),0)
            
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
    bw = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)
    
    blue_lower = np.array([  100, 100, 70])
    blue_upper = np.array([  135, 255, 255])

    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, blue_lower, blue_upper)
    
    # canny
    Threshold1 = 5;
    Threshold2 = 40;
    FilterSize = 1
    edge = cv2.Canny(bw, Threshold1, Threshold2, FilterSize)
    edge = cv2.blur(edge,(5,5))
    
    both = cv2.bitwise_and(edge, mask)
    
    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(blurred,blurred, mask=mask)
    
    contours,h = cv2.findContours(mask,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    lowest_points = []
    
    for cnt in contours:
        length = cv2.arcLength(cnt,True)
        approx = cv2.approxPolyDP(cnt,0.005*length,True)
        
        area = cv2.contourArea(cnt)
        area_per_el = area / len(approx)
        length_per_el = length / len(approx)
        if length_per_el > 50:
            points = sorted(approx, key=lambda x: x[0][1], reverse=True)
            lowest_points.append(points[0][0])
        
            cv2.drawContours(drawable_image, [approx], -1, (255, 0, 0), 2)
    
    dy = None
    if len(lowest_points) == 2:
        std_points = sorted(lowest_points, key=lambda x: x[0], reverse=True)
        x1, y1 = std_points[0]
        x2, y2 = std_points[1]
        
        dy = y2-y1
        write(drawable_image, f"dy = {dy}")        
        
        cv2.line(drawable_image,(x1,y1),(x2,y2),(0,0,255),2)
    
    return [dy, mask, edge, both]


def get_dy_edge(edge, drawable_image):
    y_left = None
    y_right = None
    n = 480
    while n > 0:
        n -= 1

        if y_left is None:
            left_edge = edge[n][:5]
            if sum(left_edge) > 0: 
                y_left = n
        if y_right is None:
            right_edge = edge[n][-5:]
            if sum(right_edge) > 0:
                y_right = n


        if y_left is not None and y_right is not None:
            break
    
    if y_left is not None and y_right is not None:
        cv2.line(drawable_image,(0,y_left),(640,y_right),(0,255,255),5)
        return y_left-y_right
    print("left, right", y_left, y_right)

while True:
    ret, frame = video.read()
    if not ret:
        video = cv2.VideoCapture(filename)
        continue
    
    dy, mask, edge, both = get_dy(frame, frame)
    dy_canny = get_dy_edge(edge, frame)
    
    write(frame, f"dy canny = {dy_canny}", 2)  
    
    cv2.imshow("frame", frame)
    cv2.imshow("alt", mask)
    cv2.imshow("edge", edge)
    
    key = cv2.waitKey(525)
    if key == 27:
        break
        

cv2.destroyAllWindows()
video.release()
```
